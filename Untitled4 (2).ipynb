```python
def generate_data_paths(data_dir):
    
    filepaths = []
    labels = []

    folds = os.listdir(data_dir)
    for fold in folds:
        if fold == 'Master Folder':
            continue
            
        foldpath = os.path.join(data_dir, fold)
        filelist = os.listdir(foldpath)
        for file in filelist:
            fpath = os.path.join(foldpath, file)
            filepaths.append(fpath)
            labels.append(fold)
            
    return filepaths, labels


filepaths, labels = generate_data_paths(data_dir)
```

```python
def create_df(filepaths, labels):

    Fseries = pd.Series(filepaths, name= 'filepaths')
    Lseries = pd.Series(labels, name='labels')
    df = pd.concat([Fseries, Lseries], axis= 1)
    return df

df = create_df(filepaths, labels)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def cat_summary_with_graph_alt(dataframe, col_name):
    # Get value counts
    counts = dataframe[col_name].value_counts()
    labels = counts.index
    values = counts.values
    
    # Create the figure and axes
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Bar plot (Countplot)
    sns.barplot(x=labels, y=values, ax=axes[0], palette='viridis')
    axes[0].set_title(f'{col_name} Countplot', fontsize=16)
    axes[0].set_xlabel('Categories', fontsize=14)
    axes[0].set_ylabel('Count', fontsize=14)
    axes[0].bar_label(axes[0].containers[0], fmt='%d', fontsize=12)  # Add count labels
    
    # Pie chart
    axes[1].pie(values, labels=labels, autopct='%1.1f%%', colors=sns.color_palette('viridis', len(labels)))
    axes[1].set_title(f'{col_name} Percentages', fontsize=16)
    
    # Display the plots
    plt.tight_layout()
    plt.show()

# Call the function
cat_summary_with_graph_alt(df, 'labels')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def plot_class_distribution(train_df, valid_df, test_df, ds_name):
    # Define the specific classes you're interested in
    specific_classes = ['Angry', 'Sad', 'Other', 'happy']
    
    # Initialize dictionaries for storing the counts for each class in each dataset
    train_counts = {cls: 0 for cls in specific_classes}
    valid_counts = {cls: 0 for cls in specific_classes}
    test_counts = {cls: 0 for cls in specific_classes}

    # Count occurrences in each dataset and assign counts
    for cls in specific_classes:
        train_counts[cls] = (train_df['labels'] == cls).sum()
        valid_counts[cls] = (valid_df['labels'] == cls).sum()
        test_counts[cls] = (test_df['labels'] == cls).sum()

    # Create a DataFrame for plotting
    class_counts_df = pd.DataFrame({
        'Class': specific_classes,
        'Training': [train_counts[cls] for cls in specific_classes],
        'Validation': [valid_counts[cls] for cls in specific_classes],
        'Testing': [test_counts[cls] for cls in specific_classes]
    })

    # Melt the DataFrame for easier plotting with seaborn
    class_counts_melted = class_counts_df.melt(id_vars=["Class"], value_vars=["Training", "Validation", "Testing"],
                                               var_name="Dataset", value_name="Count")

    # Plotting with seaborn
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Class', y='Count', hue='Dataset', data=class_counts_melted, palette="Set2")
    
    # Add labels and title
    plt.xlabel('Class', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.title(f"Class Distribution Across {ds_name}", fontsize=15)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Annotate bars with their values, positioning slightly above the bars
    for p in plt.gca().patches:
        height = p.get_height()
        plt.text(p.get_x() + p.get_width() / 2., height + 3,  # Adjusting the y-coordinate
                 f'{int(height)}', ha="center", va="bottom", fontsize=10)

    plt.show()

# Example usage
plot_class_distribution(train_df, valid_df, test_df, 'Pets Facial Expression')
```
